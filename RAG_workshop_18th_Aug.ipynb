```python
import requests #fetch pdf from url
url= "https://www.lkouniv.ac.in/site/writereaddata/siteContent/202003291621085101sanjeev_rdbms_unit-I_sql_bba_ms_4_sem.pdf"
response = requests.get(url)
```

```python
#save pdf local
with open("sql_commands.pdf", "wb") as f:
  f.write(response.content)
```

```python
from http.cookiejar import LoadError
#load document

loader = UnstructuredFileLoader("sql_commands.pdf")
documents = loader.load()
```

```python
documents
```

```python
text_splitter = CharacterTextSplitter(chunk_size=200, chunk_overlap=50)
```

```python
text_splitter
```

```python
texts= text_splitter.split_documents(documents)
```

```python
texts
```

```python
type(texts)
```

```python
texts[5]
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
persist_directory = 'vector_db'
```

```python
vectordb = Chroma.from_documents(texts, embeddings, persist_directory=persist_directory)
```

```python
retriever = vectordb.as_retriever()
```

```python
llm=ChatGroq(model= "Llama-3.1-70b-Versatile",temperature=0)
```

```python
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=retriever,return_source_documents=True)
```

```python
query= 'what are the commands and Datatypes in this pdf'
response = qa_chain.invoke({"query":query})
```
